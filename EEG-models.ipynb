```python
import pandas as pd
df = pd.read_csv(r'/Users/vishwa/Desktop/EEG/data2.csv')
```

```python
df.head()
```

```python
# Duplicates have been removed
# Considering the case 
train_data = df.iloc[:516,:]
train_data
```

```python
# since our data is not in similar range, Eg: Question No. and TRP2
# Normalizing the data

# Standard normalization using the data
# We do not need the data of everyone while processing

# NAN values have been eliminated


X = train_data.iloc[:,2:21]
Y = train_data.iloc[:,1]

```

```python
X.head()
```

```python
# Clustering into 3 ratings

#for i in Y:
    #if i==1 or i==2:
        #i = 1
    #elif i==4 or i==5:
        #i = 5
    #else:
        #i=3
```

```python
Y.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() 
# using mean/ standard deviation for normalization
X= scaler.fit_transform(X)

# till BRP 12 we have all values so will only be considering those
# first we need to split the data
# 5 class or 3 class classification - CONFIRM

# ritchieng.com/machine-learning-k-nearest-neighbors-knn/
```

```python
# Applying K-Nearest Neighbours Algorithm
# Iterating for different values of K
# Training on the entire set to find the correct value of K 
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier


knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X,Y)
Y_pred = knn.predict(X)
# to evaluate the accuracy of the model
print(metrics.accuracy_score(Y, Y_pred))


```

```python
# Applying Random Forests Classifier
# https://towardsdatascience.com/an-implementation-and-explanation-of-the-random-forest-in-python-77bf308a9b76

from sklearn.ensemble import RandomForestClassifier

# Creating a Random Foresr with 100 trees
model = RandomForestClassifier(n_estimators=100,bootstrap=True,max_features ='sqrt')

model.fit(X,Y)
```

```python
# Actual Predcitons
Y_pred = model.predict(X)
# Predictions for each class
Y_probs = model.predict_proba(X)

print(Y_pred)
print(Y_probs)
```

```python
# After Training on the entire data
# We are training on the entire data so that we can tune the parameters
print(metrics.accuracy_score(Y, Y_pred))
```

```python
# Creating a test train split

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20,random_state=10)
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
Y_probs = model.predict_proba(X_test)[:,1]
print(metrics.accuracy_score(Y_test, Y_pred))
```

```python
# Using Decision tree

rf = RandomForestClassifier( max_depth=10,random_state=7) 
rf.fit(X_train, Y_train)
sc_train = rf.score(X_train,Y_train)
```
