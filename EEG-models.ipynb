```python
import pandas as pd
df = pd.read_csv(r'/Users/vishwa/Desktop/EEG/data2.csv')
```

```python
df.head()
```

```python
# Duplicates have been removed
# Considering the case 
train_data = df.iloc[516:,:]


# because all the data is combined according to the rating
train_data = train_data.sample(frac=1).reset_index(drop=True)

train_data
```

```python
# since our data is not in similar range, Eg: Question No. and TRP2
# Normalizing the data

# Standard normalization using the data
# We do not need the data of everyone while processing

# NAN values have been eliminated


X = train_data.iloc[:,2:]
Y = train_data.iloc[:,1]

```

```python
X.drop(['Question no'],axis =1)
```

```python
Y.head()
```

```python
#from sklearn.preprocessing import StandardScaler

#scaler = StandardScaler() 
# using mean/ standard deviation for normalization
#X= scaler.fit_transform(X)

# till BRP 12 we have all values so will only be considering those
# first we need to split the data
# 5 class or 3 class classification - CONFIRM

# ritchieng.com/machine-learning-k-nearest-neighbors-knn/
```

```python
# Applying K-Nearest Neighbours Algorithm
# Iterating for different values of K
# Training on the entire set to find the correct value of K 
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier


knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X,Y)
Y_pred = knn.predict(X)
# to evaluate the accuracy of the model
print(metrics.accuracy_score(Y, Y_pred))


```

```python
# Applying Random Forests Classifier
# https://towardsdatascience.com/an-implementation-and-explanation-of-the-random-forest-in-python-77bf308a9b76

from sklearn.ensemble import RandomForestClassifier

# Creating a Random Foresr with 100 trees
model = RandomForestClassifier(n_estimators=10,bootstrap=True,max_features ='sqrt')

model.fit(X,Y)
```

```python
# Actual Predcitons
Y_pred = model.predict(X)
# Predictions for each class
Y_probs = model.predict_proba(X)

print(Y_pred)
print(Y_probs)
```

```python
Y
```

```python
# After Training on the entire data
# We are training on the entire data so that we can tune the parameters
print(metrics.accuracy_score(Y, Y_pred))
```

```python
# Creating a test train split

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20,random_state=10)
```

```python
from sklearn.neighbors import KNeighborsClassifier

train_acc = []
test_acc = []
for i in range(1,20):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,Y_train)
    acc_train = knn.score(X_train,Y_train)
    train_acc.append(acc_train)
    acc_test = knn.score(X_test,Y_test)
    test_acc.append(acc_test)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
train_score,=plt.plot(range(1,20),train_acc,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='green', markersize=5)
test_score,=plt.plot(range(1,20),test_acc,color='red',linestyle='dashed',  marker='o',
         markerfacecolor='blue', markersize=5)
plt.legend( [train_score, test_score],["Train Accuracy", "Test Accuracy"])
plt.title('Accuracy vs K neighbors')
plt.xlabel('K neighbors')
plt.ylabel('Accuracy')
```
