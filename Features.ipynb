```python
def bandpower(data, sf, band, method='welch', window_sec=None, relative=False):
    """Compute the average power of the signal x in a specific frequency band.

    Requires MNE-Python >= 0.14.

    Parameters
    ----------
    data : 1d-array
      Input signal in the time-domain.
    sf : float
      Sampling frequency of the data.
    band : list
      Lower and upper frequencies of the band of interest.
    method : string
      Periodogram method: 'welch' or 'multitaper'
    window_sec : float
      Length of each window in seconds. Useful only if method == 'welch'.
      If None, window_sec = (1 / min(band)) * 2.
    relative : boolean
      If True, return the relative power (= divided by the total power of the signal).
      If False (default), return the absolute power.

    Return
    ------
    bp : float
      Absolute or relative band power.
    """
    from scipy.signal import welch
    from scipy.integrate import simps
    from mne.time_frequency import psd_array_multitaper

    band = np.asarray(band)
    low, high = band

    # Compute the modified periodogram (Welch)
    if method == 'welch':
        if window_sec is not None:
            nperseg = window_sec * sf
        else:
            nperseg = (2 / low) * sf

        freqs, psd = welch(data, sf, nperseg=nperseg)

    elif method == 'multitaper':
        psd, freqs = psd_array_multitaper(data, sf, adaptive=True,
                                          normalization='full', verbose=0)

    # Frequency resolution
    freq_res = freqs[1] - freqs[0]

    # Find index of band in frequency vector
    idx_band = np.logical_and(freqs >= low, freqs <= high)

    # Integral approximation of the spectrum using parabola (Simpson's rule)
    bp = simps(psd[idx_band], dx=freq_res)

    if relative:
        bp /= simps(psd, dx=freq_res)
    return bp
```

```python
def features(data2):
    from scipy import signal
    import seaborn as sns
    from scipy.integrate import simps
    sns.set(font_scale=1.2)
    
    data1=[]
    #choosing the 6 frontal channels
    data1.append(data2['temp'][0])
    data1.append(data2['temp'][1])
    data1.append(data2['temp'][2])
    data1.append(data2['temp'][3])
    data1.append(data2['temp'][10])
    data1.append(data2['temp'][11])

    # Define sampling frequency and time vector
    # indices representation 0-F1,1-F2,2-F3,3-F4,4-F11,5-F12
    beta_power=[]
    beta_rel_power=[]
    alpha_power=[]
    alpha_rel_power=[]
    theta_power=[]
    theta_rel_power=[]
    beta_power_avg=[]
    beta_rel_power_avg=[]
    theta_power_avg=[]
    theta_rel_power_avg=[]
    alpha_power_avg=[]
    alpha_rel_power_avg=[]
    # our sampling rate is 1000 
    i=0
    for i in range(6):
       sf = 1000.
       data = data1[i]
       time = np.arange(data.size) / sf



    # Define window length (2 seconds)
    # our window len is 2s
       win = 2 * sf
       freqs, psd = signal.welch(data, sf, nperseg=win)

     
       # Define DELTA lower and upper limits
       low, high = 1, 3

       # Find intersecting values in frequency 
    
       idx_delta = np.logical_and(freqs >= low, freqs <= high)
       low, high = 3.5, 7.5
       idx_theta = np.logical_and(freqs >= low, freqs <= high)
       low, high =  7.5,13
       idx_alpha = np.logical_and(freqs >= low, freqs <= high)
       low, high =  14,30

    
       idx_beta = np.logical_and(freqs >= low, freqs <= high)
    # Frequency resolution
       freq_res = freqs[1] - freqs[0]  # = 1 / 4 = 0.25

    # Compute the absolute power by approximating the area under the curve
      # freq_res will be same for all cases because it considers the edge cases
       #print(idx_theta)
       #print(freq_res)

       total_power = simps(psd, dx=freq_res)
        
       beta_power.append(simps(psd[idx_beta], dx=freq_res))
       beta_rel_power.append(beta_power[i] / total_power)
    
     
       theta_power.append(simps(psd[idx_theta], dx=freq_res))
       theta_rel_power.append(theta_power[i] / total_power)
       
       alpha_power.append(simps(psd[idx_alpha], dx=freq_res))
       alpha_rel_power.append(alpha_power[i] / total_power)
       
    
       # from here we calculate all average band powers
       bpt= bandpower(data, sf, [3.5, 7.5], 'multitaper')
       bp_relt = bandpower(data, sf, [3.5, 7.5], 'multitaper', relative=True)
       theta_power_avg.append(bpt)
       theta_rel_power_avg.append(bp_relt)
       

       bpa = bandpower(data, sf, [7.5, 13], 'multitaper')
       bp_rela = bandpower(data, sf, [7.5, 13], 'multitaper', relative=True)
       alpha_power_avg.append(bpt)
       alpha_rel_power_avg.append(bp_relt)

       bpb = bandpower(data, sf, [14, 30], 'multitaper')
       bp_relb = bandpower(data, sf, [14, 30], 'multitaper', relative=True)
       beta_power_avg.append(bpt)
       beta_rel_power_avg.append(bp_relt) 
       
 
        
       final_features = (theta_rel_power+alpha_rel_power+beta_rel_power+
                        theta_power_avg + alpha_power_avg + beta_power_avg+
                        theta_rel_power_avg + alpha_rel_power_avg + beta_rel_power_avg)
    return final_features
```
